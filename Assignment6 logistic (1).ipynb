```python
#Custom binary Encoding of binary o/p variables
bank1['default']=numpy.where(bank1['default'].str.contains("yes"),1,0)
```

```python
bank1['housing']=numpy.where(bank1['housing'].str.contains('yes'),1,0)
```

```python
bank1['loan']=numpy.where(bank1['loan'].str.contains('yes'),1,0)
```

```python
bank1['Target']=numpy.where(bank1['Target'].str.contains('yes'),1,0)
```

```python
bank1
```

```python
#Replace Encoding for categorical column 'Month'
bank1['month'].value_counts()
```

```python
order={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
```

```python
bank1=bank1.replace(order)
```

```python
bank1
```

```python
bank1.info()
```

```python
#Dividing Data into i/p & o/p variables
X=pandas.concat([bank1.iloc[:,0:11],bank1.iloc[:,12:]],axis=1)
Y=bank1.iloc[:,11]
```

```python
#Logistic Regression Model
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
classifier=LogisticRegression()
classifier.fit(X,Y)
```

```python
#Model Predictions 
#predict for X dataset
y_pred=classifier.predict(X)
y_pred
```

```python
y_pred_df=pandas.DataFrame({'Actual_Y':Y,'y_pred_prob':y_pred})
```

```python
y_pred_df
```

```python
#Testing Model Accuracy
#use confusion matrix for model accuracy
from sklearn.metrics import confusion_matrix
confusion_matrix=confusion_matrix(Y,y_pred)
confusion_matrix
```

```python
#plot ROC curve & find AUC value
from sklearn.metrics import roc_auc_score
fpr,tpr,thresholds=roc_curve(Y,classifier.predict_proba(X)[:,1])
auc=roc_auc_score(Y,y_pred)

plt.plot(fpr,tpr,color='salmon',label='logit model(area=%0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
print('auc accuracy:',auc)
```
